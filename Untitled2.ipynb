```python
import gradio as gr
import random
import time

# -------------------- GLOBAL DATA -------------------- #

goals = []
marks_data = {}

motivations = [
    "🔥 Push yourself, because no one else will do it for you.",
    "🚀 Small progress is still progress.",
    "💡 Focus on improvement, not perfection.",
    "🏆 Discipline beats motivation.",
    "📈 Consistency creates success."
]

# -------------------- GOAL TRACKER -------------------- #

def add_goal(goal):
    if goal.strip() == "":
        return goals, "⚠️ Enter a valid goal!"
    goals.append({"task": goal, "done": False})
    return goals, "✅ Goal Added!"

def complete_goal(index):
    if 0 <= index < len(goals):
        goals[index]["done"] = True
    return goals

def show_goals():
    if not goals:
        return "No goals added yet."

    completed = sum(1 for g in goals if g["done"])
    total = len(goals)
    percent = int((completed / total) * 100)

    output = "<h3>🎯 Goals:</h3>"
    for i, g in enumerate(goals):
        status = "✅" if g["done"] else "❌"
        output += f"{i}. {g['task']} {status}<br>"

    output += f"<br><b>Progress:</b> {percent}%"
    return output


# -------------------- MARKS TRACKER -------------------- #

def add_marks(subject, marks):
    if subject.strip() == "":
        return "⚠️ Enter subject name."
    try:
        marks = float(marks)
    except:
        return "⚠️ Enter valid marks."

    marks_data[subject] = marks
    return "✅ Marks Added!"

def show_performance():
    if not marks_data:
        return "No marks added."

    total = sum(marks_data.values())
    avg = total / len(marks_data)

    output = "<h3>📊 Performance:</h3>"
    for sub, mark in marks_data.items():
        output += f"{sub}: {mark}<br>"

    output += f"<br><b>Average:</b> {round(avg,2)}%"
    return output


# -------------------- MOTIVATION -------------------- #

def get_motivation():
    return random.choice(motivations)


# -------------------- TIMER -------------------- #

def start_timer():
    for i in range(25, -1, -1):
        time.sleep(1)
        yield f"⏳ {i} minutes remaining"
    yield "🎉 Session Complete!"


# -------------------- UI -------------------- #

with gr.Blocks(theme=gr.themes.Soft(primary_hue="indigo")) as app:

    gr.Markdown("""
    # 🧠 Student Productivity Dashboard
    ### Build Discipline. Track Progress. Achieve More. 🚀
    """)

    with gr.Tab("🎯 Goal Tracker"):
        goal_input = gr.Textbox(label="Enter New Goal")
        add_goal_btn = gr.Button("Add Goal")
        goal_status = gr.Textbox(label="Status")
        goals_display = gr.HTML()
        refresh_goals = gr.Button("Refresh Goals")

        add_goal_btn.click(add_goal, inputs=goal_input, outputs=[goals_display, goal_status])
        refresh_goals.click(show_goals, outputs=goals_display)

    with gr.Tab("⏳ Study Timer"):
        timer_output = gr.Textbox(label="Pomodoro Timer")
        start_btn = gr.Button("Start 25 Min Timer")

        start_btn.click(start_timer, outputs=timer_output)

    with gr.Tab("📊 Marks Tracker"):
        subject_input = gr.Textbox(label="Subject")
        marks_input = gr.Textbox(label="Marks")
        add_marks_btn = gr.Button("Add Marks")
        marks_status = gr.Textbox(label="Status")
        performance_display = gr.HTML()
        show_perf_btn = gr.Button("Show Performance")

        add_marks_btn.click(add_marks, inputs=[subject_input, marks_input], outputs=marks_status)
        show_perf_btn.click(show_performance, outputs=performance_display)

    with gr.Tab("💬 Motivation"):
        motivation_text = gr.Textbox(label="Daily Motivation")
        motivate_btn = gr.Button("Get Motivation")

        motivate_btn.click(get_motivation, outputs=motivation_text)

app.launch(share=True)
```

```python
def recommend(name, user_skills, interest):

    import matplotlib.pyplot as plt

    if not name or not user_skills or not interest:
        return "Please fill all fields properly.", None

    user_skills = user_skills.lower().split(",")
    user_skills = [s.strip() for s in user_skills]

    scores = {}

    for career, data in career_db.items():
        score = 0

        match_count = 0
        for skill in user_skills:
            if skill in data["skills"]:
                match_count += 1

        skill_score = (match_count / len(data["skills"])) * 70
        score += skill_score

        if interest.lower() == data["field"].lower():
            score += 30

        scores[career] = round(score, 2)

    best_career = max(scores, key=scores.get)
    match_percentage = scores[best_career]

    result_text = f"""
Student: {name}

Recommended Career: {best_career}
Match Percentage: {match_percentage}%
Salary: {career_db[best_career]['salary']}

Roadmap:
{career_db[best_career]['roadmap']}
"""

    # CREATE FIGURE OBJECT PROPERLY
    fig, ax = plt.subplots(figsize=(8,5))
    ax.bar(scores.keys(), scores.values())
    ax.set_ylim(0, 100)
    ax.set_ylabel("Match Percentage")
    ax.set_title("Career Match Comparison")
    plt.xticks(rotation=30)

    return result_text, fig
```

```python
import gradio as gr
import pandas as pd

career_db = {
    "Data Scientist": {
        "category": "Technical",
        "skills": ["python", "machine learning", "data", "statistics"],
        "salary": "₹6–25 LPA",
        "roadmap": "Python → ML → Deep Learning → Projects"
    },
    "Software Developer": {
        "category": "Technical",
        "skills": ["java", "coding", "dsa", "programming"],
        "salary": "₹3–20 LPA",
        "roadmap": "Programming → DSA → Projects"
    },
    "Doctor": {
        "category": "Medical",
        "skills": ["biology", "medical", "diagnosis"],
        "salary": "₹8–30 LPA",
        "roadmap": "NEET → MBBS → Specialization"
    },
    "Chartered Accountant": {
        "category": "Commerce",
        "skills": ["accounts", "finance", "tax"],
        "salary": "₹7–25 LPA",
        "roadmap": "CA Foundation → CA Final"
    },
    "Graphic Designer": {
        "category": "Creative",
        "skills": ["painting", "drawing", "design", "craft"],
        "salary": "₹3–15 LPA",
        "roadmap": "Design Basics → Tools (Photoshop/Canva) → Portfolio"
    }
}

def recommend(name, skills_input):

    if not name or not skills_input:
        return "Please fill all fields properly.", None

    user_skills = [s.strip().lower() for s in skills_input.split(",")]

    scores = {}
    match_details = {}

    for career, data in career_db.items():
        matched_skills = []
        for skill in user_skills:
            if skill in data["skills"]:
                matched_skills.append(skill)

        score = (len(matched_skills) / len(data["skills"])) * 100
        scores[career] = round(score, 2)
        match_details[career] = matched_skills

    best = max(scores, key=scores.get)

    if scores[best] == 0:
        result = f"""
Student: {name}

⚠️ No strong match found.
Try adding more specific skills.
"""
    else:
        matched = ", ".join(match_details[best]) if match_details[best] else "None"
        result = f"""
Student: {name}

🎯 Recommended Career: {best}
📊 Match Percentage: {scores[best]}%
💰 Salary Range: {career_db[best]['salary']}

🗺 Roadmap:
{career_db[best]['roadmap']}

🧠 Why This Career Is Suitable?

You have {len(match_details[best])} matching skill(s):
{matched}

These skills strongly align with the core requirements of {best}.
"""

    df = pd.DataFrame({
        "Career": list(scores.keys()),
        "Match %": list(scores.values())
    })

    return result, df


# ================= UI Styling =================

custom_css = """
body {
    background: linear-gradient(135deg, #eef2ff, #e0f7fa);
    color: #111111 !important;
}

/* Blue Heading Box */
.header-box {
    background: linear-gradient(135deg, #2563eb, #3b82f6);
    color: white;
    padding: 30px;
    border-radius: 18px;
    text-align: center;
    box-shadow: 0px 10px 30px rgba(0,0,0,0.25);
    margin-bottom: 30px;
}

/* Full width section box */
.section-box {
    background: white;
    padding: 30px;
    border-radius: 18px;
    box-shadow: 0px 8px 25px rgba(0,0,0,0.1);
    margin-bottom: 30px;
}

/* Stylish Button */
button {
    background: linear-gradient(45deg, #1d4ed8, #2563eb) !important;
    color: white !important;
    font-weight: 600 !important;
    border-radius: 12px !important;
    padding: 12px !important;
    box-shadow: 0px 6px 18px rgba(37,99,235,0.5) !important;
    transition: 0.3s ease-in-out;
}

button:hover {
    transform: translateY(-3px);
    box-shadow: 0px 10px 25px rgba(37,99,235,0.7) !important;
}

textarea {
    border-radius: 10px !important;
}
"""

with gr.Blocks(theme=gr.themes.Soft(primary_hue="indigo"), css=custom_css) as app:

    gr.Markdown("""
    <div class="header-box">
        <h1 style="font-weight:900; font-size:34px;">
            🤖 Smart AI Career Recommender System
        </h1>
        <h3>Skill-Based Career Analysis with Explanation 📊</h3>
    </div>
    """)

    with gr.Column(elem_classes="section-box"):

        name = gr.Textbox(label="👤 Enter Your Name")
        skills = gr.Textbox(label="🛠 Enter Your Skills (comma separated)")
        btn = gr.Button("🚀 Analyze Career")
        output_text = gr.Textbox(label="📄 Career Recommendation", lines=14)

    gr.Markdown("## 📊 Career Match Comparison")

    output_chart = gr.BarPlot(x="Career", y="Match %", height=450)

    btn.click(
        recommend,
        inputs=[name, skills],
        outputs=[output_text, output_chart]
    )

app.launch(share=True)
```
